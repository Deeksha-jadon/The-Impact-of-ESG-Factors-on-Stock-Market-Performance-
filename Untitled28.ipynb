```python
# Group by sector and calculate the average annual stock return for 2024.
# This helps us see which sectors are performing best (or worst) this year.
sector_summary = df.groupby('Sector')['Annual Return (%) 2024'].mean().reset_index()

# Plot the bar graph
plt.figure(figsize=(12, 6))
sns.barplot(x='Annual Return (%) 2024', y='Sector', data=sector_summary, palette='coolwarm')
plt.title('Average Stock Return by Sector')
plt.xlabel('Annual Return (%) 2024')
plt.ylabel('Sector')
plt.tight_layout()
plt.show()
```

```python
# Group by sector and calculate average stock return
sector_summary = df.groupby('Sector')['company'].count().reset_index()

# Plot the bar graph
plt.figure(figsize=(12, 6))
sns.barplot(x='company', y='Sector', data=sector_summary, palette='coolwarm')
plt.title('company count by Sector')
plt.xlabel('company')
plt.ylabel('Sector')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = ['2020', '2021', '2022', '2023', '2024']
correlations = [-0.1376, 0.1667, 0.3021, 0.1038, -0.128166]

# Convert correlation values to a NumPy array 
import numpy as np
correlations_np = np.array(correlations)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(years, correlations_np, marker='o', linestyle='-', color='skyblue')

# Add horizontal reference line
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# Titles and labels
plt.title('Correlation Between ESG Scores and Annual Returns (2020-2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Correlation Coefficient')
plt.ylim(-0.2, 0.4)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Select relevant columns and drop missing values
data = df[['company', 'esg_risk_score_2024', 'Annual Return (%) 2024']].dropna()

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data[['esg_risk_score_2024', 'Annual Return (%) 2024']])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# Plot clusters with company names
plt.figure(figsize=(12, 7))
sns.scatterplot(data=data,
                x='esg_risk_score_2024',
                y='Annual Return (%) 2024',
                hue='Cluster',
                palette='Set2',
                s=100)

# Annotate company names
for i in range(data.shape[0]):
    plt.text(data['esg_risk_score_2024'].iloc[i] + 0.5,
             data['Annual Return (%) 2024'].iloc[i],
             data['company'].iloc[i],
             fontsize=8)

plt.title('KMeans Clustering: ESG Risk Score vs Stock Return (with company Names)')
plt.xlabel('ESG Risk Score (2024)')
plt.ylabel('Annual Return (%) (2024)')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```
